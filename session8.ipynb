```python
# Number dividing by 0
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
    print("C=",c)
except ZeroDivisionError:
    print("You are trying to divide a number by 0")    
```

# Index error 

```python
lst=[10,2,3,6,5,4,8,8]
print(lst[10])
    
```

```python
try:
    lst=[10,20,30,40,50,60,70,80,90]
    indexnumber=int(input("Enter your index number "))
    print(lst[indexnumber])
except IndexError:
    print("List index out of range")
```

# type error 

```python
a='sujit bondre'
b=23
c=a+b
print("C=",c)
```

```python
try:
    a='sujit bondre'
    b=23
    c=a+b
    print("C=",c)
except TypeError as e:
    print(e)
```

# Raise Keyword 


```python
try:
    age=int(input("Enter your age: "))
    if age>=18:
        raise TypeError("Person is eligible for voting")
    else:
        print("Person is not eligible for voting")
except TypeError as e:
        print(e)
```

# finally Keyword 

```python
try:
    a=int(input("Enter a number :"))
    b=int(input("Enter second number : "))
    c=a+b
    print("c=",c)
except ValueError as e:
    print(e)
finally:
    print("finally your calculation is completed")
```

```python
# Password validation using Exception handling
```

```python
u="Sujit"
p="123654"
try:
    username=input("Enter your username: ")
    password=input("Enter your password: ")
    if(username==u and password==p):
        print("User logged in successfully")
    else:
        raise ValueError("Wrong username or password")
except ValueError as e:
    print(e)
```
