```python
def print_message():
    print("Welcome to India")
```

```python
import threading
```

```python
t1=threading.Thread(target=print_message)
```

```python
t1.start()
```

```python
def calculate_square(n):
    print(f"The square of number {n} is {n*n}")
```

```python
t2=threading.Thread(target=calculate_square,args=(5,))
t2.start()
```

```python
l1=[10,20,3,5,6,9,8,7,4,50]
```

```python
for i in l1:
   t2=threading.Thread(target=calculate_square,args=(i,))
   t2.start()
```

```python
import urllib.request
def download_file(lnk):
    file_name=lnk.split("/")[-1]
    print("file is downloading........")
    path="C:/Users/sujit/Downloads/"+file_name
    urllib.request.urlretrieve(lnk,path)
    print(f"{file_name} is downloading")
```
